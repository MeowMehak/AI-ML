```python
# 1. Student details dictionary
student = {"name": "Mehak", "age": 20, "course": "BCA"}
print(student)

```

```python
# 2. Access values
print(student["name"])
```

```python
# 3. Add new key-value

student["year"] = 3
print(student["year"])
```

```python
# 4. Update value
student["age"] = 21
print(student["age"] )
```

```python
# 5. Delete an item
del student["year"]
print(student)
```

```python
# 6. Loop through keys and values
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
# 7. Character frequency
s = "hello"
freq = {}
for c in s:
    freq[c] = freq.get(c, 0) + 1
print(freq)
```

```python
# 8. Merge two dictionaries
d1 = {"a": 1}
d2 = {"b": 2}
merged = {**d1, **d2}
print(merged)

```

```python
# 9. Check if key exists
print("name" in student)
```

```python
# 10. Sort by keys
print(dict(sorted(student.items())))
```
