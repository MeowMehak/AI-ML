```python
# 1. Print 1 to 100
for i in range(1, 101):
    print(i)
```

```python
# 2. Even numbers between 1 and 50
for i in range(2, 51, 2):
    print(i)
```

```python
# 3. Sum from 1 to 100
total = 0
for i in range(1, 101):
    total += i
print(total)  
```

```python
# 4. Multiplication table
n = 5
for i in range(1, 11):
    print(f"{n} x {i} = {n * i}")
```

```python
# 5. Count vowels using loop
s = "Hello World"
count = 0
for c in s.lower():
    if c in "aeiou":
        count += 1
print(count)
```

```python
# 6. Factorial using loop
n = 5
result = 1
for i in range(1, n + 1):
    result *= i
print(result)

```

```python
# 7. Pattern using nested loops
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
# 8. Reverse a number using loop
n = 12345
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print(rev)

```

```python
# 9. Largest number in list using loop
lst = [3, 1, 7, 2, 9]
largest = lst[0]
for x in lst:
    if x > largest:
        largest = x
print(largest)
```

```python
# 10. Check palindrome using loop
s = "racecar"
is_pal = True
for i in range(len(s) // 2):
    if s[i] != s[-(i + 1)]:
        is_pal = False
        break
print(is_pal)
```
