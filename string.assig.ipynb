```python
s = "hello world"

# 1. Reverse
print(s[::-1])

# 2. Check palindrome
word = "racecar"
print(word == word[::-1])

# 3. Count vowels
print(sum(1 for c in s if c in "aeiou"))

# 4. Uppercase
print(s.upper())

# 5. Replace word
print(s.replace("world", "Python"))

# 6. Count words
print(len(s.split()))

# 7. Character frequency
freq = {}
for c in s:
    freq[c] = freq.get(c, 0) + 1
print(freq)

# 8. Remove spaces
print(s.replace(" ", ""))

# 9. Check anagram
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)
print(is_anagram("listen", "silent"))  # True

# 10. Find duplicate characters
seen, dupes = set(), set()
for c in s:
    if c in seen:
        dupes.add(c)
    seen.add(c)
print(dupes)
```
