```python
num = int(input("Enter a number: "))

count = 0

while num > 0:
    num = num // 10
    count += 1

print("Number of digits =", count)
```

```python
num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number =", reverse)
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print(factorial(5))
```

```python
def prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(prime(7))
```

```python
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

```python
def count_vowels(text):
    count = 0

    for ch in text.lower():
        if ch in "aeiou":
            count += 1

    return count

print(count_vowels("Education"))
```

```python
def even_numbers(lst):
    even = []

    for i in lst:
        if i % 2 == 0:
            even.append(i)

    return even

print(even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# 1. Add two numbers
def add(a, b):
    return a + b

# 2. Factorial
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

# 3. Check prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# 4. Reverse a string
def reverse_string(s):
    return s[::-1]

# 5. Count vowels
def count_vowels(s):
    return sum(1 for c in s.lower() if c in "aeiou")

# 6. Maximum in a list
def find_max(lst):
    return max(lst)

# 7. Square of a number
def square(n):
    return n ** 2

# 8. Check palindrome
def is_palindrome(s):
    return s == s[::-1]

# 9. Return even numbers from list
def get_evens(lst):
    return [x for x in lst if x % 2 == 0]

# 10. Sum of list elements
def list_sum(lst):
    return sum(lst)
```
