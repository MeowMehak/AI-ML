```python
lst = [4, 1, 7, 2, 9, 2, 5]

# 1. Largest element
print(max(lst))

```

```python
# 2. Smallest element
print(min(lst))
```

```python
# 3. Remove duplicates
print(list(set(lst)))
```

```python
# 4. Reverse a list
print(lst[::-1])
```

```python
# 5. Sort ascending
print(sorted(lst))
```

```python
# 6. Sum of elements
print(sum(lst))
```

```python
# 7. Count occurrence
print(lst.count(2))
```

```python
# 8. Merge two lists
a = [1, 2, 3]
b = [4, 5, 6]
print(a + b)
```

```python
# 9. Second largest
unique = list(set(lst))
unique.sort()
print(unique[-2])
```

```python
# 10. Split even and odd
evens = [x for x in lst if x % 2 == 0]
odds  = [x for x in lst if x % 2 != 0]
print("Even:", evens, "| Odd:", odds)
```
