```python
t = (10, 20, 30, 20, 40)

# 1. Create and print
print(t)

# 2. Access elements
print(t[0], t[-1])

# 3. Length
print(len(t))

# 4. Count occurrence
print(t.count(20))

# 5. Index of element
print(t.index(30))

# 6. Convert to list
print(list(t))

# 7. Convert list to tuple
lst = [1, 2, 3]
print(tuple(lst))

# 8. Maximum value
print(max(t))

# 9. Minimum value
print(min(t))

# 10. Concatenate two tuples
t2 = (50, 60)
print(t + t2)
```
