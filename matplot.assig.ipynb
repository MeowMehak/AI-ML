```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Line Chart

students = ["Aman", "Riya", "Rahul", "Sneha"]
marks = [80, 90, 75, 85]

plt.figure(figsize=(7, 4))
plt.plot(students, marks, marker='o', color='steelblue', linewidth=2)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.tight_layout()
plt.savefig("q1_line_chart.png", dpi=120)
plt.show()



```

```python
# Bar Chart

brands = ["Samsung", "Apple", "MI", "OnePlus", "Vivo"]
sales = [50, 40, 70, 30, 60]

plt.figure(figsize=(7, 4))
plt.bar(brands, sales, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7'])
plt.title("Mobile Brand Sales")
plt.xlabel("Brand")
plt.ylabel("Sales (units)")
plt.tight_layout()
plt.savefig("q2_bar_chart.png", dpi=120)
plt.show()
```

```python
# Pie Chart

activities = ["Study", "Sleep", "Games", "Exercise"]
hours = [6, 8, 5, 2]

plt.figure(figsize=(6, 6))
plt.pie(hours, labels=activities, autopct='%1.1f%%',
        colors=['#FF9999', '#66B2FF', '#99FF99', '#FFD700'],
        startangle=140, explode=(0.05, 0.05, 0.05, 0.05))
plt.title("Daily Activity Distribution")
plt.tight_layout()
plt.savefig("q3_pie_chart.png", dpi=120)
plt.show()
```

```python
# Scatter Plot

hours = [1, 2, 3, 4, 5, 6]
marks = [30, 40, 50, 60, 70, 85]

plt.figure(figsize=(7, 4))
plt.scatter(hours, marks, color='darkorange', s=100, edgecolors='black', zorder=3)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("q4_scatter_plot.png", dpi=120)
plt.show()

# ───────────────────────────────
```

```python
#  Multiple Line Graph

days = [1, 2, 3, 4, 5]
aman = [60, 65, 70, 75, 80]
riya = [55, 60, 68, 72, 78]

plt.figure(figsize=(7, 4))
plt.plot(days, aman, marker='o', linestyle='-',  color='royalblue',  label='Aman', linewidth=2)
plt.plot(days, riya, marker='s', linestyle='--', color='tomato',     label='Riya', linewidth=2)
plt.title("Student Progress Over Days")
plt.xlabel("Day")
plt.ylabel("Score")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.4)
plt.tight_layout()
plt.savefig("q5_multi_line.png", dpi=120)
plt.show()
```

```python
# Histogram

marks_data = [45, 56, 67, 78, 89, 90, 56, 67, 76, 88, 92]

plt.figure(figsize=(7, 4))
plt.hist(marks_data, bins=5, color='mediumpurple', edgecolor='white', linewidth=1.2)
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("q6_histogram.png", dpi=120)
plt.show()
```

```python
#  Customized Chart

subjects = ["Math", "Science", "English", "History", "Art"]
scores   = [88, 92, 76, 84, 95]

plt.figure(figsize=(8, 5))
plt.bar(subjects, scores,
        color=['#E74C3C','#3498DB','#2ECC71','#F39C12','#9B59B6'],
        edgecolor='black', linewidth=0.8)
plt.plot(subjects, scores, marker='^', color='black', linewidth=1.5,
         linestyle='--', label='Trend')
plt.title("Subject-wise Score Analysis", fontsize=14, fontweight='bold', pad=12)
plt.xlabel("Subject", fontsize=11)
plt.ylabel("Score", fontsize=11)
plt.ylim(60, 105)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.savefig("q7_custom_chart.png", dpi=120)
plt.show()
```

```python
# Question 8 — IPL Data Visualization
# ─────────────────────────────────────────────
players = ["Virat", "Rohit", "Dhoni", "Gill"]
runs    = [82, 75, 50, 90]

colors = ['#FF6B00', '#004BA0', '#FFD700', '#007DC5']

plt.figure(figsize=(7, 5))
bars = plt.bar(players, runs, color=colors, edgecolor='black', linewidth=0.8, width=0.5)
for bar, run in zip(bars, runs):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1,
             str(run), ha='center', va='bottom', fontweight='bold', fontsize=11)
plt.title("IPL Player Runs", fontsize=14, fontweight='bold')
plt.xlabel("Player")
plt.ylabel("Runs")
plt.ylim(0, 110)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig("q8_ipl_chart.png", dpi=120)
plt.show()
```

```python
# Question 9 — Weather Analysis
# ─────────────────────────────────────────────
days_week    = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature  = [30, 32, 35, 33, 31, 29, 28]

plt.figure(figsize=(8, 4))
plt.plot(days_week, temperature, marker='o', color='orangered',
         linewidth=2.5, markersize=8, markerfacecolor='white',
         markeredgewidth=2)
plt.fill_between(days_week, temperature, alpha=0.15, color='orangered')
plt.title("Weekly Temperature Analysis", fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week")
plt.ylabel("Temperature (°C)")
plt.ylim(24, 38)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("q9_weather_analysis.png", dpi=120)
plt.show()
```
